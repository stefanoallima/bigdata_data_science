```
# Esta funcion elimina la puntuaci贸n del texto que recibe como argumento
def removePunctuation(text):
  import re
  return re.sub(r'[^A-Za-z0-9 ]', '', text).lower().strip()

fileName = '/databricks-datasets/cs100/lab1/data-001/shakespeare.txt'

```

```
# 1. Cargue el archivo localizado en fileName como un RDD llamado shakespeareRDD.
shakespeareRDD = sc.textFile(fileName);
shakespeareRDD.take(1005)
```

```
# 2. Aplique la funci贸n removePunctuation al contenido de shakespeareRDD. Use map.
shakespeareRDD = shakespeareRDD.map(removePunctuation);
shakespeareRDD.take(1005)
```

```
shakespeareRDDD = shakespeareRDD.map(lambda x: removePunctuation(x))
shakespeareRDDD.take(1005)
```

```
# 3. Transforme shakespeareRDD a shakespeareWordsRDD separando las palabras por espacios. Use map con la funci贸n split(' ')
shakespeareWordsRDD = shakespeareRDD.flatMap(lambda shakespeareRDD:shakespeareRDD.split(" "));
#shakespeareWordsRDD.take(100)
```

```
# 4. Elimine las palabras de longitud 0 de shakespeareWordsRDD. Guarde el resultado en wordsRDD. Use filter.
wordsRDD = shakespeareWordsRDD.filter(lambda shakespeareWordsRDD: len(shakespeareWordsRDD)>0)
#wordsRDD.collect()
```

```
# 5. Imprima el conteo de palabras en wordsRDD. Use count
print(wordsRDD.count())
```

```
# 6. Calcule la frecuencia de cada palabra en wordsRDD y guarde el resultado en freqRDD. Use map y reduceByKey.
freqRDD = wordsRDD.map(lambda word: (word, 1)).reduceByKey(lambda a, b: a + b)
```

```
# 7. Imprima el contenido de freqRDD. Use collect y display
display(freqRDD.collect())
```
